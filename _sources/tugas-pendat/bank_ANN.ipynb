```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/Rifqi-05/pendata/main/Bank_Personal_Loan_Modelling.csv')
df.head()
```

```python
X = df.drop(['CreditCard'], axis=1)
y = df['CreditCard']
```

```python
from sklearn.preprocessing import StandardScaler
X_std=StandardScaler().fit_transform(X)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
# Memisahkan fitur dan target
X = df.drop('CreditCard', axis=1)
y = df['CreditCard']
```

```python
# Standarisasi fitur
scaler = StandardScaler()
X_std = scaler.fit_transform(X)
```

```python
# Memisahkan data menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(X_std, y, test_size=0.1, random_state=42)
```

```python
# Membuat objek MLPClassifier
mlp = MLPClassifier(hidden_layer_sizes=(100,100,100), max_iter=1000, alpha=0.0001,
                     solver='sgd', verbose=10,  random_state=21,tol=0.001)
```

```python
# Melatih model dengan data latih
mlp.fit(X_train, y_train)
```

```python
# Memprediksi kelas dari data uji
y_pred = mlp.predict(X_test)
```

```python
# Menghitung akurasi
accuracy = accuracy_score(y_test, y_pred)
print("Akurasi:", accuracy)
```

```python
# Menyimpan model menggunakan pickle
model_filename = 'loan.pkl'
with open(model_filename, 'wb') as file:
    pickle.dump(mlp, file)

print("Model saved as", model_filename)
```
