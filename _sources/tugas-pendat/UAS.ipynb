# **Perhitungan Naive Bayes tanpa PCA**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/Rifqi-05/pendata/main/Bank_Personal_Loan_Modelling.csv')
dataset.head(10)
```

```python
dataset =dataset
X = dataset.iloc[:,:6].values
y = dataset['CreditCard'].values
dataset.head(5)
```

```python
from sklearn.preprocessing import StandardScaler
X_std=StandardScaler().fit_transform(X)
X_std
```

```python
df_std = pd.DataFrame(X_std, columns=dataset.columns[:6])
df_std.to_csv('preprocessed_data.csv', index=False)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.05, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test) 
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
from sklearn.metrics import accuracy_score 
print ("Accuracy : ", accuracy_score(y_test, y_pred))
cm
```

```python
# Mendefinisikan fungsi untuk memprediksi RiskLevel dari input pengguna
def predict_risk_level(input_data):
    # Lakukan preprocessing pada input data
    input_data = sc.transform([input_data])  # Lakukan penskalaan pada input data

    # Lakukan prediksi menggunakan model Gaussian Naive Bayes
    predicted_risk_level = classifier.predict(input_data)

    return predicted_risk_level
```

```python
# Contoh penggunaan fungsi predict_risk_level dengan input pengguna
input_user = [15,120,80,7.01,98,70]
predicted_risk = predict_risk_level(input_user)
print("Predicted Risk Level:", predicted_risk)
```

# **Perhitungan KNN tanpa PCA**

```python
import pandas as pd
import sklearn
```

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/Rifqi-05/pendata/main/Bank_Personal_Loan_Modelling.csv')
dataset.head(10)
```

```python
dataset = dataset
X = dataset.iloc[:,:6].values
y = dataset['CreditCard'].values
dataset.head(5)
```

```python
from sklearn.preprocessing import StandardScaler
X_std=StandardScaler().fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test) 
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
from sklearn.metrics import accuracy_score 
print ("Accuracy : ", accuracy_score(y_test, y_pred))
cm
```

```python
# Memisahkan data menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(X_std, y, test_size=0.1, random_state=42)
```

```python
# Memprediksi kelas dari data uji
y_pred = mlp.predict(X_test)
```

```python
# Menghitung akurasi
accuracy = accuracy_score(y_test, y_pred)
print("Akurasi:", accuracy)
```
